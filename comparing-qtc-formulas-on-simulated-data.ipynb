```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style = "darkgrid")
```

```python
hr = list(range(55, 180, 5)) # defining a heart rate range from 50 to 180 beats per minute
qt = list(range(400, 300, -4))
'''
60 bpm == 1000 ms
120 bpm == 500 ms
180 bpm == 333 ms
'''
rr = []
for i in hr:
    rr.append(round(1000 / ( 60 / i )))
# for i in range(len(hr)): print(hr[i], qt[i], rr[i])
qtcBaz, qtcFri, qtcHod = [], [], []
for i in range(len(hr)):
    qtcBaz.append(np.round(qt[i] / np.sqrt(1000 / rr[i])))
    qtcFri.append(np.round(qt[i] / np.cbrt(1000 / rr[i])))
    qtcHod.append(np.round(qt[i] + ( 1.75 * (hr[i] - 60))))
for i in range(len(hr)): print(hr[i], qt[i], rr[i], qtcBaz[i], qtcFri[i], qtcHod[i])
```

```python
plt.title("Comparing QTc Formulas")
plt.legend(["Bazette", "Friderica", "Hodges"])
sns.lineplot(x=hr, y=qtcBaz)
sns.lineplot(x=hr, y=qtcFri)
sns.lineplot(x=hr, y=qtcHod)
plt.legend(('Baz', 'Fri', 'Hod'))
plt.xlabel("heart rate")
plt.ylabel("corrected qt duration")
```
